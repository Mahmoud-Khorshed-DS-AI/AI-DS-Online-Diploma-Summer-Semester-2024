## Visualization 5: Count Plot - Sales by Month:
Visualizing the number of sales made in each month using a count plot.

```python
E_comm_USA.info()
```

```python
# Visualization 5: Count Plot - Sales by Month 
E_comm_USA['Order Date'] = pd.to_datetime(E_comm_USA['Order Date']) 
E_comm_USA.info()
```

```python
E_comm_USA.sample()
```

```python
E_comm_USA['Month'] = E_comm_USA['Order Date'].dt.month # Feature engineering
E_comm_USA.sample(5)
```

```python
plt.figure(figsize=(5, 3))
sns.countplot(x='Month', data=E_comm_USA)  # univarite
plt.xlabel('Month') 
plt.ylabel('Sales Count') 
plt.title('Sales by Month') 
plt.show()
```


## Visualization 7: Line Plot - Monthly Sales Trend:

## Visualization 8: Clustered Bar Chart - Sales by Customer Type and Product Category:

 Visualizing the sales breakdown by customer type and product category using a Clustered bar chart.

```python
E_comm_USA.sample(2)
```

```python
customer_category_sales = E_comm_USA.pivot_table(index='Segment',
                                        columns='Category',
                                        values='Sales',
                                        aggfunc='sum')
customer_category_sales
```

```python
customer_category_sales.plot(kind='bar', stacked=False, figsize=(10, 6)) 

plt.xlabel('Customer Type') 
plt.ylabel('Total Sales') 
plt.title('Sales by Customer Type and Product Category') 
plt.legend(title='Product Category', bbox_to_anchor=(1, 1)) 
plt.show()
```
